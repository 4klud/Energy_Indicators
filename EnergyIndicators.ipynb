```python
import pandas as pd
import numpy as np
x = pd.ExcelFile('EnergyIndicators.xlsx')
energy = x.parse(skiprows=1)
energy = energy[['Country','Petajoules','Gigajoules','%']]
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy[['Energy Supply', 
        'Energy Supply per Capita', 
        '% Renewable']] = energy[['Energy Supply',
                                  'Energy Supply per Capita',
                                  '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)
energy['Energy Supply'] = energy['Energy Supply']*1000000
energy['Country'] = energy['Country'].replace({'China, Hong Kong Special Administrative Region':'Hong Kong',
                                               'United Kingdom of Great Britain and Northern Ireland19':'United Kingdom',
                                               'Republic of Korea':'South Korea',
                                               'United States of America':'United States',
                                               'Iran (Islamic Republic of)':'Iran',
                                               'Switzerland17': 'Switzerland',
                                               'Ukraine18': 'Ukraine',
                                               'The former Yugoslav Republic of Macedonia' : 'North Macedonia'})
energy['Country'] = energy['Country'].str.replace(r" \(.*\)","")
energy = energy.set_index('Country', drop = True)
energy
```
